```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Se lee el archivo csv Hechos
hechos = pd.read_csv('Hechos.csv')
```

```python
# Especifico la ruta del archivo excel
homicidios = 'Censo 2022.xlsx'

# Especifico el nombre de la hoja a leer
nombre_de_hoja = 'Cuadro 1.1'

# Utilizo la funcion de pandas para leer la hoja específica del excel
Censo_2022 = pd.read_excel(homicidios, sheet_name=nombre_de_hoja)

# Se almacena la poblacion total de CABA censada en el 2022
poblacion_total = int(Censo_2022['Unnamed: 3'][3])
```

```python
hechos.dtypes
```

```python
# Se castea la columna fecha de tipo objeto a tipo datetime
hechos['fecha'] = pd.to_datetime(hechos['fecha'], format='%Y-%m-%d', errors='coerce')
```

```python
hechos['fecha']
```

```python
# Se crea la columna semestre para su posterior uso en la medida del KPI
hechos['semestre'] = hechos['fecha'].dt.month.map({1: '1er', 2: '1er', 3: '1er', 4: '1er',
                                                   5: '1er', 6: '1er', 7: '2do', 8: '2do',
                                                   9: '2do', 10: '2do', 11: '2do', 12: '2do'}) + ' semestre'
```

```python
hechos
```

```python
# Se agrupa por las columnas 'año' y 'semestre', y suma la columna 'n_victimas'
siniestros_Xsemestre = hechos.groupby(['aaaa', 'semestre'])['n_victimas'].sum().reset_index()
```

```python
# Se calcula la tasa de siniestros por semestre para cada año usando como población total el dato obtenido del Censo 2022
siniestros_Xsemestre['tasa_siniestros'] = round((siniestros_Xsemestre['n_victimas'] / poblacion_total) * 100000,2)
```

```python
# Se genera la columna 'tasa_siniestros_anterior' utilizando el metodo shift para colocarlos una fila por debajo
siniestros_Xsemestre['tasa_siniestros_anterior'] = siniestros_Xsemestre['tasa_siniestros'].shift()
```

```python
# Se calcula la variación porcentual con los datos calculados para 'tasa_siniestros' y 'tasa_siniestros_anterior'
siniestros_Xsemestre['variacion_porcentual'] = round(siniestros_Xsemestre['tasa_siniestros'].pct_change() * 100,2)
```

```python
# Se renombra la columna que corresponde al año
siniestros_Xsemestre = siniestros_Xsemestre.rename(columns={'aaaa': 'año'})
```
